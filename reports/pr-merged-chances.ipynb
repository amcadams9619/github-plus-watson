```python
import os
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as pyp
```

```python
pd.set_option('max_columns', 50)
#Inline imaging
%matplotlib inline
df = pd.read_csv('pr_report_closed.csv', delimiter='|')
df.set_index('pr_id', inplace=True)
df.head(10)
```

```python
#Merged and Closed
prs_closed = df[df['pr_status'] == 'closed']

pr_clmrg = df[df['is_merged'] == "merged"]

pr_open = df[df['pr_status'] == "open"]

total = sum(pr_clmrg['num_comments'])
nums = len(pr_clmrg['num_comments'])
```

```python
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

data = [go.Bar(
            x=df.index,
            y=df.values
    )]
layout = go.Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)',
    font=dict(size=18, color='#FFFFFF'),
    width=1000,
    height=1000
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
